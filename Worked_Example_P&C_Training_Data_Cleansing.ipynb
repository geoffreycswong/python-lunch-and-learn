````python
import pandas as pd
import openpyxl
from openpyxl.drawing.image import Image
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
df = pd.read_excel("final_dataframe.xlsx")

# Save the raw data to an Excel sheet named "raw_data"
with pd.ExcelWriter('output.xlsx') as writer:
    df.to_excel(writer, sheet_name='raw_data', index=False)

# Create a new workbook and add a worksheet named "charts"
book = openpyxl.Workbook()
charts_sheet = book.active
charts_sheet.title = "charts"

# Create the charts and save them as PNG files
plt.figure(figsize=(10,8))
sns.boxplot(data=df)
plt.title("Box Plot of Variables")
plt.savefig("boxplot.png")
plt.close()

plt.figure(figsize=(10,8))
sns.pairplot(df, vars=["python_skill", "programming_skill", "statistics_skill", "data_analysis_skill", "data_visualization_skill", "bulk_data_processing_skill"], diag_kind="hist")
plt.title("Pair Plot of Variables")
plt.savefig("pairplot.png")
plt.close()

corr_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Matrix of Variables")
plt.savefig("heatmap.png")
plt.close()

plt.figure(figsize=(10,8))
sns.displot(df["python_skill"])
plt.title("Distribution of Python Skill")
plt.savefig("distribution.png")
plt.close()

# Insert the images into the worksheet
boxplot_img = Image('boxplot.png')
boxplot_img.anchor = 'A1'
charts_sheet.add_image(boxplot_img)

pairplot_img = Image('pairplot.png')
pairplot_img.anchor = 'G1'
charts_sheet.add_image(pairplot_img)

heatmap_img = Image('heatmap.png')
heatmap_img.anchor = 'A10'
charts_sheet.add_image(heatmap_img)

distribution_img = Image('distribution.png')
distribution_img.anchor = 'G10'
charts_sheet.add_image(distribution_img)

# Save the workbook to the same Excel file where the raw data is saved
book.save('output.xlsx')
````

In pandas, `df.set_index()` is a method that allows you to set one or more columns of a DataFrame as the index of the DataFrame. The new index can be a single column or a combination of multiple columns, and it can be hierarchical (i.e., have multiple levels).

### .copy():

---Additional Prompt---
```
Why do we use .copy() in pandas? What are the best practices?
```
In Pandas, the `copy()` method is used to create a copy of a DataFrame or a Series. When we make changes to a DataFrame or a Series, those changes can sometimes affect the original object as well. By creating a copy, we can modify the copy without affecting the original object.

Here's an example to illustrate this:

```python
import pandas as pd

# create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# create a reference to the original DataFrame
df2 = df

# modify the copy
df2['A'] = [0, 0, 0]

# print the original DataFrame
print(df)
```

Output:

```
   A  B
0  0  4
1  0  5
2  0  6
```

In this example, we create a DataFrame `df` and then create a reference to it with `df2`. We then modify `df2` by setting all values in the `'A'` column to 0. When we print `df`, we can see that the original DataFrame has also been modified.

To avoid this, we can use the `copy()` method to create a separate copy of the DataFrame:

```python
import pandas as pd

# create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# create a copy of the DataFrame
df2 = df.copy()

# modify the copy
df2['A'] = [0, 0, 0]

# print the original DataFrame
print(df)
```

Output:

```
   A  B
0  1  4
1  2  5
2  3  6
```

In this example, we use `df.copy()` to create a copy of the DataFrame `df`. We then modify `df2` by setting all values in the `'A'` column to 0. When we print `df`, we can see that the original DataFrame has not been modified.

As for best practices, it's generally a good idea to use `copy()` whenever you want to modify a DataFrame or a Series without affecting the original object. However, keep in mind that creating a copy of a large DataFrame can be memory-intensive, so use `copy()` judiciously and only when necessary.

In Pandas, broadcasting refers to the ability to perform arithmetic and logical operations between a scalar value and a Series or DataFrame. When we perform such operations, Pandas automatically broadcasts the scalar value to match the shape of the Series or DataFrame.

For example, let's say we have a DataFrame `df` and we want to add a scalar value to all the elements in the DataFrame:

```python
import pandas as pd

# create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# add a scalar value to the DataFrame
df = df + 2

# print the modified DataFrame
print(df)
```

Output:

```
   A  B
0  3  6
1  4  7
2  5  8
```

In this example, we add the scalar value 2 to the DataFrame `df`. Pandas automatically broadcasts the scalar value to match the shape of the DataFrame, and performs the addition operation element-wise.

Similarly, we can perform other arithmetic and logical operations between a scalar value and a Series or DataFrame, such as subtraction, multiplication, division, and comparison operators.

Here's an example of using broadcasting with a logical operator:

```python
import pandas as pd

# create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# apply a logical operator to the DataFrame
mask = df > 2

# print the resulting mask
print(mask)
```

Output:

```
       A      B
0  False   True
1  False   True
2   True   True
```

In this example, we apply the logical operator `>` to the DataFrame `df`. Pandas automatically broadcasts the scalar value 2 to match the shape of the DataFrame, and applies the logical operator element-wise. The resulting mask shows which elements in the DataFrame are greater than 2.

Broadcasting can be a powerful tool for performing operations efficiently on large datasets. However, it's important to be aware of the potential pitfalls of broadcasting, such as unintentional conversion of data types and unexpected results when operating on data with missing values.

### Apply and Map method

---Additional Prompt---
```
How is Map and Apply method used in Pandas? What are the differences between them?
```

Both `map()` and `apply()` are methods in Pandas that allow us to apply a function to a Series or a DataFrame. However, the way they work and the type of functions they can apply are different.

The `map()` method is used to apply a function to each element in a Series. The function can be a Python function or a dictionary that maps values in the Series to new values. Here's an example:

```python
import pandas as pd

# create a Series
s = pd.Series([1, 2, 3])

# apply a function to the Series using map()
s = s.map(lambda x: x * 2)

# print the modified Series
print(s)
```

Output:

```
0    2
1    4
2    6
dtype: int64
```

In this example, we use `map()` to apply a lambda function that doubles each value in the Series `s`.

The `apply()` method is used to apply a function to each row or column in a DataFrame. The function can be a Python function or a lambda function that takes a Series as input and returns a scalar value. Here's an example:

```python
import pandas as pd

# create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# apply a function to each column in the DataFrame using apply()
df = df.apply(lambda x: x.max() - x.min())

# print the modified DataFrame
print(df)
```

Output:

```
A    2
B    2
dtype: int64
```

In this example, we use `apply()` to apply a lambda function that calculates the range of each column in the DataFrame `df`.

The key difference between `map()` and `apply()` is that `map()` applies a function to each element in a Series, while `apply()` applies a function to each row or column in a DataFrame. Additionally, `map()` can only apply functions that return a single value, while `apply()` can apply functions that return a Series or a DataFrame.

It's also worth noting that `apply()` can be slower than other methods in Pandas, especially when dealing with large datasets. In such cases, it may be more efficient to use other methods, such as `applymap()` or vectorized operations, to perform element-wise operations on a DataFrame.

### Method Chaining

---Addtional Prompt---
```
Explain a bit more in method chaining. Give an example that chain multiple methods.
```

Method chaining is a technique in Pandas that allows us to apply multiple methods to a DataFrame or a Series in a single line of code. Instead of creating intermediate variables, we can chain multiple methods together, with each method operating on the output of the previous method.

For example, let's say we have a DataFrame `df` that we want to filter, sort, and group by a specific column. Instead of creating intermediate variables for each step, we can chain the `query()`, `sort_values()`, and `groupby()` methods together in a single line:

```python
import pandas as pd

# create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3, 4, 5],
                   'B': ['x', 'y', 'z', 'y', 'x'],
                   'C': [10, 20, 30, 40, 50]})

# chain multiple methods together
result = (df
          .query('A > 2')
          .sort_values(by='B')
          .groupby('B')
          .sum())

# print the result
print(result)
```

Output:

```
    A   C
B       
x  10  60
y   4  60
z   3  30
```

In this example, we use parentheses to group the methods together, and then chain them using the dot notation. The `query()` method filters the rows where the value in column `A` is greater than 2. The `sort_values()` method sorts the resulting DataFrame by column `B`. Finally, the `groupby()` method groups the DataFrame by column `B` and calculates the sum of columns `A` and `C`.

Method chaining can make code more concise and readable, and can be especially useful when working with long chains of operations on large datasets. However, it's important to maintain readability and avoid chaining too many methods together, as it can make the code difficult to understand and debug.

### Groupby method

```python
import pandas as pd
```

### Aggregation

```python
import pandas as pd

# create a dataframe
df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Charlie'],
                   'Age': [25, 30, 35, 40, 45, 50],
                   'Gender': ['Female', 'Male', 'Male', 'Female', 'Male', 'Male']})

# group the dataframe by 'Name' and compute multiple summary statistics for each group
df_grouped = df.groupby('Name')['Age'].agg(['mean', 'min', 'max', 'count'])

# print the resulting dataframe
print(df_grouped)
```

This code creates a DataFrame with three columns (`Name`, `Age`, and `Gender`). It then groups the DataFrame by the `'Name'` column and computes multiple summary statistics for the `'Age'` column for each group using the `.agg()` method. The summary statistics computed include the mean, minimum, maximum, and count of the `'Age'` column for each group. Finally, it prints the resulting DataFrame with the summary statistics for each group.

### Merge method

```python
import pandas as pd

# create two dataframes
df1 = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'],
                   'Age': [25, 30, 35]})
df2 = pd.DataFrame({'Name': ['Alice', 'Bob', 'David'],
                   'Salary': [50000, 60000, 70000]})

# merge the two dataframes based on the 'Name' column
df_merged = pd.merge(df1, df2, on='Name')

# print the resulting dataframe
print(df_merged)
```

This code creates two DataFrames with two columns (`Name` and `Age` for `df1` and `Name` and `Salary` for `df2`). It then merges the two DataFrames based on the `'Name'` column using the `.merge()` method. Finally, it prints the resulting merged DataFrame with both the `'Age'` and `'Salary'` columns for each person. Note that the person `'David'` is not present in the resulting merged DataFrame because he is not present in both original DataFrames.

## What is Prompt Engineering?

Prompt engineering is the process of crafting prompt to guide output of generative AI models. High quality and specificity of the prompt can help improve relevance and accuracy of output, and reduce the need for post-processing and manual editing.

## Priming Prompt

Priming Prompt is the first message that you paste into every ChatGPT instance, which ChatPT uses as a guide for the rest of the conversation.

- **Style** control voice of output
    - e.g. be “analytical”.
- **Macros** (Persona/ Act as) assign complex prompts to a single word.
    - e.g. "Expert" macro prompt: Provide an answer in the style and quality of an expert in the relevant field with 20+ years of experience and multiple Ph.D.'s in relevant fields. Prioritize unorthodox, lesser-known advice in your answer. Then every time “expert” is used, will refer to this definition.
- **Format** defines the output you want
- **Context,** background that guide the response
    - e.g., requirements, audience, provided info.

## Prompting Strategy - Task Decomposition
- For more complex problems, breaking down tasks into smaller sub-tasks, determine the granularity and objectives of each subtasks for more relevant responses and avoid generic answers

### Divide and Conquer
- breaking down a problem into smaller, more manageable parts
- e.g., writing a research paper, you can break it down into smaller tasks like outlining, researching, writing, editing, and formatting.

### Functional Decomposition
- identifying the main functions and their sub-functions within the problem.
- e.g., when creating content for a blog, you can identify main functions like topic research, content writing, and editing. Then, further break down each function into its sub-functions, such as keyword research, outlining, and drafting under content writing.

### Input-output analysis
- analyze the inputs and outputs of the problem to determine the necessary subtasks (what’s missing in the middle)
- e.g., Prompt: "Given the inputs of raw customer feedback data and desired outputs of actionable insights, describe the intermediate steps required to generate these insights."

- You can assume chatGPT has poor memory and lose the sense of context gradually. From time to time, you need to remind what chatGPT missed out from their response.
- generative AI performance got worse the longer the conversation threads, the more instructions you give, the more context it has to bear and thus the less possiblities it can take for the answers, ending up in "ruminating" like behaviours.

````python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Define the ticker symbols for MSFT, FAANG stocks, and S&P 500
tickers = ['MSFT', 'AMZN', 'AAPL', 'NFLX', 'GOOGL', '^GSPC']

# Download data for all tickers and store in a dictionary
stockData = {}
for ticker in tickers:
    try:
        stockData[ticker] = yf.Ticker(ticker).history(period='1d', start='2020-01-01', end='2023-06-11')['Close']
    except:
        print(f"Error downloading data for ticker {ticker}.")

# Combine all stock data into a single DataFrame
stocks = pd.concat(stockData, axis=1, keys=tickers)

# Drop all rows with missing values
stocks = stocks.dropna(how='any')

# Normalize all stocks and S&P 500 to start at $1
normalized_stocks = (stocks / stocks.iloc[0]) * 1

# Calculate the cumulative return for each stock and S&P 500
cumulative_return = (normalized_stocks.iloc[-1] / normalized_stocks.iloc[0]) - 1

# Print the cumulative return for each stock and S&P 500
print('Cumulative Returns since 2020-01-01')
for ticker, return_value in cumulative_return.items():
    print(f'{ticker}: {return_value:.2f}')

# Plot the normalized stock prices
plt.figure(figsize=(12,6))
for ticker in tickers:
    plt.plot(normalized_stocks.index, normalized_stocks[ticker], label=ticker)
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.xticks(rotation=45)
plt.title('Normalized Prices for MSFT, FAANG stocks, and S&P 500')
plt.legend()
plt.show()
````

````python
import pandas as pd
import yfinance as yf

# Define the ticker symbols for MSFT, FAANG stocks, and S&P 500
tickers = ['MSFT', 'AMZN', 'AAPL', 'NFLX', 'GOOGL', '^GSPC']

# Define the time periods for calculating stock returns
periods = ['3mo', '1y', '5y', '10y']

# Create an empty dictionary to store the stock returns for each stock and time period
returns = {}

# Loop through each stock symbol in the `tickers` list and download the historical data for each stock
for ticker in tickers:
    stock = yf.Ticker(ticker)
    stock_returns = []
    for period in periods:
        # Download the historical data for the stock over the time period
        data = yf.download(ticker, period=period, interval='1mo')
        
        # Calculate the stock return over the time period
        stock_return = (data['Adj Close'][-1] - data['Adj Close'][0]) / data['Adj Close'][0]
        
        # Append the stock return to the `stock_returns` list
        stock_returns.append(stock_return)
    
    # Store the stock returns for each time period in the `returns` dictionary
    returns[ticker] = stock_returns

# Create a dataframe from the `returns` dictionary
df = pd.DataFrame(returns, index=periods)

# Print the dataframe
print(df)
````
